```python
red_wine_df = pd.read_csv('../resources/data/wine-quality/winequality-red.csv', delimiter=';')
```

```python
white_wine_df = pd.read_csv('../resources/data/wine-quality/winequality-white.csv', delimiter=';')
```

```python
wine_df = pd.concat([red_wine_df, white_wine_df])
```

```python
wine_arr = wine_df.to_numpy()
```

```python
wine_arr.shape
```

```python
wine_t = torch.from_numpy(wine_arr)
```

```python
X_origin, t_origin = wine_t[:, :-1], wine_t[:, -1].to(torch.int64)
```

```python
X_mean = torch.mean(X_origin, dim=0)
```

```python
X_var = torch.var(X_origin, dim=0)
```

```python
X_norm = (X_origin - X_mean) / torch.sqrt(X_var)
```
